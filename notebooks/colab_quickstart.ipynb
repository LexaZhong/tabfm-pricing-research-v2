## 3. Run the benchmark via the hosted TabPFN API (`tabpfn-client`)

The local `tabpfn` package needs a GPU. On a **CPU-only** runtime it errors with
*"Running on CPU with more than 1000 samples is not allowed"*. `tabpfn-client`
sidesteps this by running inference on Prior Labs' GPU servers — no local GPU
required.

**Data caveat:** the client uploads your feature rows to an external service.
**Public datasets only** (freMTPL2freq is public). Never use it with internal
company data — `run_benchmark.py` refuses `--internal-csv` + `tabpfn-client`.

Run the login cell **once** (free Prior Labs account; it caches an access token
to disk that the benchmark subprocess reuses), then run the benchmark cell.

_If you did enable a T4 GPU above, you can instead use the local model with
`--models tabpfn` and skip the login cell._

````python
# --n-test kept small: each test row is a hosted API prediction, so a full
# 100k test set would be slow and may hit free-tier rate limits.
!python src/run_benchmark.py --models tabpfn-client --sample-sizes 2000 --seeds 42 --n-test 10000
````
